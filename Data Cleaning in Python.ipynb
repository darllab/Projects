Why do we need to clean our data?

 - to ensure that the data insights we've come up with are accurate.

You may not want to present it to the stakeholders uncleaned, inaccurate, or whatever. So you have to do data cleaning!

The dataset I use is intended to be uncleaned so that I can show you the process of data cleaning. This is just a small dataset, so it will be easy for us to do data cleaning. :)

```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Import dataset. You may have encountered different file. But it's okay. Pandas can still read it. 

df = pd.read_excel(r'/Users/darllabulagner/Downloads/uncleaned bike sales data.xlsx')
df
```

```python
pd.set_option('display.max.rows', 90)
```

```python
df
```

```python
# Drop duplicates if any.

df.drop_duplicates()
```

```python
# Let's ensure that there's no duplicate at all. 

df.loc[df.duplicated()]
```

```python
# Let's check if there's rows that is NULL. 

df.isna()
```

```python
# This is also another way of checking the number of null values. 

df.isna().sum()
```

```python
# Now that we already identified null values. Let's have an analysis

df.dropna(subset=['Day'], inplace=True)
df.dropna(subset=['Age_Group'], inplace=True)
df.dropna(subset=['Product_Description'], inplace=True)
df.dropna(subset=['Order_Quantity'], inplace=True)
```

```python
df
```

```python
# Now that we already drop null values. 
# I have noticed that the columns 'Day' and 'Order Quantity' is in float or decimal.
# Let's change it to integer.


df['Day'] = df['Day'].astype(int)
df['Order_Quantity'] = df['Order_Quantity'].astype(int)
```

```python
df
```

```python
# I also want to remove the column 'Day', 'Month', 'Year' because for me it's redundant

df = df.drop(columns = "Day")
df = df.drop(columns = "Month")
df = df.drop(columns = "Year")
```

```python
df
```

```python
# And I think that's it! Now we export our dataset. To do this..
# Note that you can do this in any file type such as Excel, CSV, JSON etc.


df.to_csv('processed_data.csv')
```
